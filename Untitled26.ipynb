```python
import pandas as pd  # Import Pandas
import matplotlib.pyplot as plt  # Import Matplotlib

# Read the Titanic dataset
df = pd.read_csv('C:/Users/anton/Downloads/Titanic-Dataset 1.csv')

# Create a pie chart of the 'Embarked' column
plt.pie(df['Embarked'].value_counts(), labels=df['Embarked'].value_counts().index, autopct='%1.1f%%')

# Add a title
plt.title('Distribution of Embarked Passengers')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('C:/Users/anton/Downloads/Titanic-Dataset 1.csv')

# Create bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x=df["Pclass"], y=df["Age"], estimator=pd.Series.max, ci=None)

# Add labels and title
plt.xlabel("Passenger Class")
plt.ylabel("Average Age")
plt.title("Average Age for Each Passenger Class")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('C:/Users/anton/Downloads/shopping_trends.csv')

# Count number of males and females
gender_counts = df['Gender'].value_counts()

# Bar Chart for Gender

sns.barplot(x=gender_counts.index, y=gender_counts.values, palette=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Number of Male and Female Passengers')
plt.show()
```
